# Handling Missing Values

```python
df.isnull().sum()
```

## Percentage of Missing Values

```python
percent_missing = df.isnull().sum() * 100 / len(df)
missing_value_df = pd.DataFrame({'column_name': df.columns,
                                 'percent_missing': percent_missing}).round(2).sort_values(by='percent_missing',ascending=False).reset_index(drop=True)
missing_value_df
```

## Impute or replace missing values

```python
df2 = df.copy()
```

```python
# Use mode to fill missing values in categorical columns
df2['Jersey_Number'] = df2['Jersey_Number'].fillna(df['Jersey_Number'].mode()[0])
df2['Nation_Position'] = df2['Nation_Position'].fillna(df['Nation_Position'].mode()[0])
df2['Nation_Jersey_Number'] = df2['Nation_Jersey_Number'].fillna(df['Nation_Jersey_Number'].mode()[0])
df2['Team_Position'] = df2['Team_Position'].fillna(df['Team_Position'].mode()[0])

# Replace with 'N/A' because assumptions can lead to inaccuracies in the data
df2['Joined_Date'] = df2['Joined_Date'].fillna("N/A")
df2['Contract_Expiry_Date'] = df2['Contract_Expiry_Date'].fillna("N/A")

# Use mean to fill missing values in numerical columns
df2['Physic'] = df2['Physic'].fillna(df['Physic'].mean())
df2['Defending'] = df2['Defending'].fillna(df['Defending'].mean())
df2['Dribbling'] = df2['Dribbling'].fillna(df['Dribbling'].mean())
df2['Passing'] = df2['Passing'].fillna(df['Passing'].mean())
df2['Shooting'] = df2['Shooting'].fillna(df['Shooting'].mean())
df2['Pace'] = df2['Pace'].fillna(df['Pace'].mean())
```

```python
df2.count()
```

```python
## computing values in Ball Control and Stamina columns
df2['Ball_Control'] = [eval(expression) for expression in df2['Ball_Control']]

df2['Ball_Control']
```

```python
df2['Stamina'] = [eval(expression) for expression in df2['Stamina']]

df2['Stamina']
```

# Bivariate Analysis (using two variables)

## Catplot (discrete vs continuous)

```python
plt.figure(figsize=(6,8))
colors = ['red', 'green','blue']
sns.boxplot(data =df2, y='Stamina', x='Body_Type',palette=colors)
plt.title('Stamina and Body type')
plt.show()
```

##### Lean and Normal body types show almost similar stamina distributions, with median stamina around 65-70. Stocky players have a slightly lower median stamina but show greater variability. There are a few outliers in all three groups, particularly for players with very low or very high stamina. 

## Scatter plot (continuous vs continuous)

##### The scatter plot displays the relationship between players' shooting and passing abilities.There is a positive correlation between the two variables, meaning that as a player's shooting skill improves, their passing ability improves as well. The correlation is however not strong. The points show some spread, suggesting that there still is variability.

```python
sns.scatterplot(data=df2, x="Shooting", y="Passing")
```

## Correlation plot

```python
plt.figure(figsize = (15,9))
sns.heatmap(round(df2.corr(numeric_only=True),2),cmap='YlGnBu',annot=True)
```

##### Strong Positive correlation (>0.7) between Overall Rating and Potential Rating, Weight_kg and Height_cm, Shooting and Dribbling, Passing and Dribbling, Ball control and skill moves. No strong negative correlation (<-0.7) between any variables.

# Multivariate

```python
df2['Body_Type']= df2['Body_Type'].astype('category')
df2['Preferred_Foot']= df2['Preferred_Foot'].astype('category')
df2['Skill_Moves']= df2['Skill_Moves'].astype('category')


sns.catplot(x="Skill_Moves", hue="Body_Type", col="Preferred_Foot",
                data=df2, kind="count",
                height=5, aspect=1);
```

##### For both left-footed and right-footed players, the most common skill moves rating is 2, followed by 3. Right-footed players have a higher count in all skill move categories compared to left-footed players. Across both foot preferences, Normal body types have a greater count than Lean and Stocky types in all skill move categories.

# Grouping and Aggregation

```python
df2.groupby(['Club'])['Overall_Rating'].mean().nlargest(20).round(2).reset_index(name="Avg_Rating")
```

##### Players at FC Barcelona have the highest average overall rating. Among the top 20 clubs with the highest average player ratings, scores range consistently between 73 and 79, with no significant drops in performance.

```python
df2.groupby(['Body_Type','Skill_Moves']).agg({'Pace': ['mean', 'min', 'max']}).round(2)
```

##### Players with a Normal body type and Skill moves rating 5 have the highest mean pace. Players with Stocky body type and skill moves rating 2 have the lowest mean pace. A pattern that can be observed across all body types is that skill moves rating 2 has the minimum pace and skill moves rating 3 has the maximum pace. 

```python
df2.groupby(['Weight_Kg'])['Stamina'].mean().round(2).nlargest(20).reset_index(name="Avg_Stamina")
```

##### The highest mean stamina is for players with 51 kgs weight. There is a significant dip between the highest and second highest stamina. The highest mean value is 96 and the next 19 values for mean stamina are within the range 64-68.
