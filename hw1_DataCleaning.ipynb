## Cleaning Data

### Checking for null values

```python
df33.isnull().sum()
```

```python
# Replace NaN values in the access code, email, region, city, Q4, Q5, Q6, Q7

# all values in access code and email address are null so it's better to drop them.
df33 = df33.drop('Access Code', axis=1)
df33 = df33.drop('Email Address', axis=1)
```

```python
# can not assume someone's region or city so median, mode etc can not be used here 
df33['Region'] = df33['Region'].fillna('Not Available')
df33['City / Town / District'] = df33['City / Town / District'].fillna('Not Available')
```

```python
# changing data type of height then replacing nan values with mean value as height is a numeric value and mean will
# be a good approximation
df33['Q4'] = df33['Q4'].str.replace(' ', '')
df33['Q4'] = pd.to_numeric(df33['Q4'], errors='coerce')
df33['Q4'] = df33['Q4'].fillna(df33['Q4'].mean()) 
```

```python
# using mode for Country of citizenship, undergraduate degree, expected graduation data as as it reflects the most common characteristic in that category
# and  helps maintain the distribution of categories in the dataset.

df33['Q5'] = df33['Q5'].fillna(df33['Q5'].mode()[0])
df33['Q6'] = df33['Q6'].fillna(df33['Q6'].mode()[0])
df33['Q7'] = df33['Q7'].fillna(df33['Q7'].mode()[0])
```

```python
# can not use mean, median, mode as this is an open ended question and assumptions can introduce bias or misrepresentation.
df33['Q15'] = df33['Q15'].replace('', 'No Response')
df33.head()
```

### Checking for whitespaces

```python
for col in df33.columns:    
    if df33[col].dtype == 'object':
        df33[col] = df33[col].str.strip()
    else:
        pass

df33.head()
```

### Capitalizing the first letter of each word

```python
for col in df33.columns:    
    if df33[col].dtype == 'object':
        df33[col] = df33[col].str.title()
    else:
        pass

df33.head()
```
